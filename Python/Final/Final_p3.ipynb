```python
import numpy as np

```

```python

A = np.array([[2, 4, 6], [1, 3, 5], [7, 9, 11]])
B = np.array([[8, 5, 2], [7, 4, 1], [6, 3, 0]])

```

```python

sum_matrix = A + B

```

```python

product_matrix = A * B

```

```python

dot_product = np.dot(A, B.T)

```

```python

determinant_A = np.linalg.det(A)

print(f"Element-wise Sum:\n{sum_matrix}")
print(f"Element-wise Product:\n{product_matrix}")
print(f"Dot Product:\n{dot_product}")
print(f"Determinant of A: {determinant_A}")
```
